```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('csv_files/headbrain.csv')
```

```python
X = data['Head Size(cm^3)'].values
```

```python
y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
```

```python
mean_y = np.mean(y)
```

```python
m = np.sum((X - mean_x) * (y - mean_y)) / np.sum((X - mean_x) ** 2)
```

```python
b = mean_y - m * mean_x
```

```python
predicted_y = m * X + b
```

```python
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X, predicted_y, color="red", label="Regression Line")
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.show()
```

```python
TSS = np.sum((y - mean_y) ** 2)
RSS = np.sum((y - predicted_y) ** 2)  
R_squared = 1 - (RSS / TSS)
```

```python
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
print(f"R-squared value:{R_squared}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('csv_files/headbrain.csv')
```

```python
X = df['Head Size(cm^3)'].values.reshape(-1, 1)
```

```python
y = df['Brain Weight(grams)'].values
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
y_pred = model.predict(X)
```

```python
beta1 = model.coef_[0]
beta0 = model.intercept_
r2 = r2_score(y, y_pred)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.show()
```

```python
print(f"Slope (1): {beta1}")
print(f"Intercept (0): {beta0}")
print(f"R-squared Value: {r2}")
```
